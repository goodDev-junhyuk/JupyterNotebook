# 01. Matplotlib_Pie_Chart.


```
# Pie_Chart는 범주별 구성 비율을 원형으로 나타내는 그래프
# pie() 함수 이용

import matplotlib.pyplot as plt

ratio = [34, 32, 16, 18]
labels = ['Apple', 'Banana', 'Melon', 'Grapes']
explode = [0, 0.10, 0, 0.10]
colors = ['silver', 'gold', '#8fd9b6', '#d395d0']

# 중심에서 벗어나는 정도 설정 explode 이용, shadow를 이용하여 그림자 표현
# colors로 다양한 색상 지정, Hex Code로도 지정 가능.
plt.pie(ratio, labels=labels, autopct='%.1f%%', 
        startangle=260, counterclock=False, 
        explode=explode, shadow=True, colors=colors)
plt.show()
```

# 03. 여러 그래프 그리기

```
import matplotlib.pyplot as plt
import numpy as np

x1 = np.linspace(0.0, 5.0)
x2 = np.linspace(0.0, 2.0)

y1 = np.cos(2 * np.pi * x1) * np.exp(-x1)
y2 = np.cos(2 * np.pi * x2)

plt.subplot(2, 1, 1)
plt.plot(x1, y1, 'o-')
plt.title('1st Graph')
plt.ylabel('Damped oscillation')

plt.subplot(2, 1, 2)
plt.plot(x2, y2, '.-')
plt.title('2nd Graph')
plt.xlabel('time (s)')
plt.ylabel('Undamped')

plt.tight_layout()
plt.show()
```

# 04. Matplotlib 컬러맵 설정



```
import matplotlib.pyplot as plt
import numpy as np

np.random.seed(0)
arr = np.random.standard_normal((8, 100))

# subplot 함수를 사용하여 네 영역에 각각의 그래프 표현.

plt.subplot(2, 2, 1)
#plt.scatter(arr[0], arr[1], c=arr[1], cmap = 'spring')
plt.scatter(arr[0], arr[1], c=arr[1])
plt.spring()
plt.title('spring')
plt.colorbar()

plt.subplot(2, 2, 2)
plt.scatter(arr[2], arr[3], c=arr[3])
plt.summer()
plt.title('summer')
plt.colorbar()

plt.subplot(2, 2, 3)
plt.scatter(arr[4], arr[5], c=arr[5])
plt.autumn()
plt.title('autumn')
plt.colorbar()

plt.subplot(2, 2, 4)
plt.scatter(arr[6], arr[7], c=arr[7])
plt.winter()
plt.title('winter')
plt.colorbar()

plt.tight_layout()
plt.show()
```

# 05. Matplotlib 텍스트 삽입 및 스타일 지정.

```
import matplotlib.pyplot as plt
import numpy as np

a = 2.0 * np.random.randn(10000) + 1.0
b = np.random.standard_normal(10000)
c = 20.0 * np.random.rand(5000) - 10.0

# 각각의 폰트스타일 지정
font1 = {'family' : 'serif',
         'color' : 'darkred',
         'weight' : 'normal',
         'size' : 16}

font2 = {'family' : 'Times New Roman',
        'color' : 'blue',
        'weight' : 'bold',
        'size' : '12',
        'alpha' : 0.7}

font3 = {'family' : 'Arial',
         'color' : 'forestgreen',
         'style' : 'italic',
         'size' : 14}
    

plt.hist(a, bins=100, density=True, alpha=0.7, histtype='step')
# 설정한 Font스타일 추가.
plt.text(1.0, 0.35, 'np.random.randn()', fontdict=font1)
plt.hist(b, bins=50, density=True, alpha=0.5, histtype='stepfilled')
plt.text(2.0, 0.20, 'np.random.standard_normal()', fontdict=font2)
plt.hist(c, bins=100, density=True, alpha=0.9, histtype='step')
plt.text(5.0, 0.08, 'np.random.rand()', fontdict=font3)
plt.show()
```

# 06. Matplotlib 수학적 표현 사용


```
import matplotlib.pyplot as plt

plt.plot([1, 2, 3, 4], [1, 4, 9, 16])
plt.xlabel('X-Label')
plt.ylabel('Y-Label')

# 'r'은 파이썬 문자열을 raw String으로 표현하도록 한다.
# Matplotlib의 수학적 표현은 두 개의 달러 기호($) 사이에 위치해야 한다.
# 위(Superscripts) 첨자, 아래(Subscripts) 첨자 추가.
plt.text(1, 15, r'$\alpha^2 > \beta_5$', fontdict={'size' : 16})

plt.show()
```

```
import matplotlib.pyplot as plt

plt.plot([1, 2, 3, 4], [1, 4, 9, 16])
plt.xlabel('X-Label')
plt.ylabel('Y-Label')

# 분수 표현
plt.text(1, 15, r'$\frac{1}{2} + \frac{3}{4} = \frac{5}{4}$', fontdict={'size':16})

plt.show()
```

```
import matplotlib.pyplot as plt

plt.plot([1, 2, 3, 4], [1, 4, 9, 16])
plt.xlabel('X-Label')
plt.ylabel('Y-Label')

#근호 및 제곱근 표현
plt.text(1, 15, r'$\sqrt{2} + \sqrt[3]{x} = y$', fontdict={'size': 16})

plt.show()
```

```
import matplotlib.pyplot as plt

plt.plot([1, 2, 3, 4], [1, 4, 9, 16])
plt.xlabel('X-Label')
plt.ylabel('Y-Label')

# 다양한 연산자 사용.
plt.text(1, 15, r'$\sin (x)\ \cos (x) \ \tan (x)$', fontdict={'size': 16})
plt.text(1, 12, r'$\lim_{x\rightarrow 2} (x^2 - x + 2)$', fontdict={'size': 16})
plt.text(1, 8, r'$\sum_{n=0}^{10}{(n^2 + n)}$', fontdict={'size': 16})

plt.show()
```

# Matplotlib 객체 지향 인터페이스2


```
import numpy as np
import matplotlib.pyplot as plt

x = np.arange(1, 5) # [1, 2, 3, 4]

fig, ax = plt.subplots(2, 2, sharex=True, sharey=True, squeeze=True)
ax[0][0].plot(x, np.sqrt(x)) # left-top
ax[0][1].plot(x, x) # right-top
ax[1][0].plot(x, -x+5) # left-bottom
ax[1][1].plot(x, np.sqrt(-x+5)) # right-bottom

plt.show()
```
